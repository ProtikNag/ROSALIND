# Problem BA1A

```python
import re

def get_overlapped_occurances(text, pattern):
    count = len(re.findall('(?='+pattern+')', text))
    return count
```

```python
text = input()
pattern = input()

print("The substring appeared :", get_overlapped_occurances(text, pattern), " times.")
```

# Problem BA1B

```python
def get_all_k_mers(pattern, k):
    l = list()
    length = len(pattern)
    for i in range(0,length-k+1):
        s = pattern[i:i+k]
        l.append(s)
        
    return l
```

```python
def most_freq_k_mers(text, k):
    l = get_all_k_mers(text, k)
    l = list(set(l))
    occurances = [get_overlapped_occurances(text, x) for x in l]

    maxi = max(occurances)
    maxi_list = [x for x in range(0,len(occurances)) if occurances[x]==maxi]
    
    ans = list()

    for i in maxi_list:
        ans.append(l[i])
        
    return ans
```

```python
text = input()
k = int(input())

l = most_freq_k_mers(text, k)
for i in l:
    print(i, end=' ')
```

# Problem BA1C

```python
def reverse_complement(text):
    ans = ""
    for i in range(len(text)-1, -1, -1):
        if(text[i]=='A'):
            ans += 'T'
        elif(text[i]=='T'):
            ans += 'A'
        elif(text[i]=='C'):
            ans += 'G'
        elif(text[i]=='G'):
            ans += 'C'
    return ans
```

```python
text = input()
ans = reverse_complement(text)
print(ans)
```

# Problem BA1D

```python
def get_overlapped_occurances_position(text, pattern):
    ans = list()
    for i in range(0,len(text)-len(pattern)+1):
        s = text[i:i+len(pattern)]
        if(s==pattern):
            ans.append(i)
    return ans
```

```python
pattern = input()
text = input()

ans = get_overlapped_occurances_position(text, pattern)

for i in ans:
    print(i, end=' ')
```

# Problem BA1E

```python
def find_all_lt_clumps(genome,k,l,t):
    l_mers = get_all_k_mers(genome, l)
    ans = list()
    for x in l_mers:
        d = dict()
        k_mers = get_all_k_mers(x, k)
        for y in k_mers:
            if(not y in d):
                d[y] = 1
            elif(y in d):
                d[y] += 1
        for key,value in d.items():
            if(value>=t):
                ans.append(key)
            
    return list(set(ans))
```

```python
genome = input()
k = int(input())
l = int(input())
t = int(input())

ans = find_all_lt_clumps(genome, k, l, t)

for i in ans:
    print(i, end=' ')
```

# Problem BA1F

```python
def skew(genome):
    ans = list()
    ans.append(0)
    cnt = 0
    for i in genome:
        if(i=='C'):
            cnt -= 1
            ans.append(cnt)
        elif(i=='G'):
            cnt += 1
            ans.append(cnt)
        else:
            ans.append(cnt)
    return ans
```

```python
genome = input()

ans = skew(genome)
mini = min(ans)

for i in range(0,len(ans)):
    if(ans[i]==mini):
        print(i, end=' ')
```

# Problem BA1G

```python
def hamming_dis(a,b):
    cnt = 0
    for i in range(0, len(a)):
        if(a[i]!=b[i]):
            cnt += 1
    return cnt
```

```python
dna1 = input()
dna2 = input()

print(hamming_dis(dna1, dna2))
```

# Problem BA1H

```python
def approximate_matches(pattern, text, d):
    k_mers = get_all_k_mers(text, len(pattern))
    ans = list()
    for i in range(0, len(k_mers)):
        x = k_mers[i]
        if(hamming_dis(x,pattern)<=d):
            ans.append(i)
    return ans
```

```python
pattern = input()
text = input()
d = int(input())

ans = approximate_matches(pattern, text, d)
for i in ans:
    print(i, end=' ')
```

# Problem BA1I

```python
l = ['A', 'C', 'G', 'T']

def recursion(i,k,ans,st):
    if(i==k):
        st = st + " " + ans
        return st
    for x in range(0,4):
        st = recursion(i+1, k, ans+l[x], st)
    return st
    
def gen_all_k_mers(i,k,ans):
    x = recursion(i, k, ans, "")
    ret = x.split()
    return ret
```

```python
def most_freq_words_with_mismatch(text, k, d):
    all_possible_k_mers = gen_all_k_mers(0, k, "")
    matches = []
    for i in all_possible_k_mers:
        x = approximate_matches(i, text, d)
        matches.append(max(len(x),0))
    maxi = max(matches)
    ans = []
    for i in range(0, len(matches)):
        x = matches[i]
        if(x==maxi):
            ans.append(all_possible_k_mers[i])
    return ans
```

```python
text = input()
k = int(input())
d = int(input())

ans = most_freq_words_with_mismatch(text, k, d)

for i in ans:
    print(i, end=' ')
```

# Problem BA1J

```python
def most_freq_words_with_mismatch_comp(text, k, d):
    all_possible_k_mers = gen_all_k_mers(0, k, "")
    matches = []
    for i in all_possible_k_mers:
        x1 = approximate_matches(i, text, d)
        x2 = approximate_matches(reverse_complement(i), text, d)
        matches.append(max(len(x1)+len(x2),0))
    maxi = max(matches)
    ans = []
    for i in range(0, len(matches)):
        x = matches[i]
        if(x==maxi):
            ans.append(all_possible_k_mers[i])
    return ans
```

```python
text = input()
k = int(input())
d = int(input())

ans = most_freq_words_with_mismatch_comp(text, k, d)

for i in ans:
    print(i, end=' ')
```

# Problem BA1K

```python
def gen_freq_array(dna, k):
    k_mers = gen_all_k_mers(0, k, "")
    k_mers.sort()
    ans = []
    for x in k_mers:
        i = get_overlapped_occurances(dna, x)
        ans.append(i)
    return ans
```

```python
text = input()
k = int(input())

ans = gen_freq_array(text, k)

for i in ans:
    print(i, end=' ')
```

# Problem BA1L

```python
def patterntonumber(pat):
    d = {'A': 0, 'C': 1, 'G': 2, 'T': 3}
    ans = 0
    length = len(pat)-1
    for i in range(0, len(pat)):
        x = length-i
        y = d[pat[i]]
        ans += (y*(4**x))
    return ans
```

```python
dna = input()
print(patterntonumber(dna))
```

# Problem  BA1M

```python
numb_to_pat = {0:'A', 1:'C', 2:'G', 3:'T'}

def num_to_pattern(i, k):
    if(k==1):
        return numb_to_pat[i]

    pref_ind = i/4
    rem = i%4
    pref_pat = num_to_pattern(int(pref_ind), k-1)
    ans = pref_pat+numb_to_pat[rem]
    return ans
```

```python
index = int(input())
k = int(input())

print(num_to_pattern(index, k))
```

# Problem BA1N

```python
def neighbor(pattern, d):
    if(d==0):
        l = list()
        l.append(pattern)
        return l
    if(len(pattern)==1):
        return ['A', 'C', 'G', 'T']
                 
    n = []
    next_door = neighbor(pattern[1:],d)
                 
    for x in next_door:
        if(hamming_dis(pattern[1:],x)<d):
            for y in ['A', 'C', 'G', 'T']:
                n.append(y+x)
        else:
            n.append(pattern[0]+x)
    return n
```

```python
text = input()
d = int(input())

ans = neighbor(text, d)

for i in ans:
    print(i, end='\n')
```
