# Problem BA3A

```python
def get_all_k_mers(pattern, k):
    l = list()
    length = len(pattern)
    for i in range(0,length-k+1):
        s = pattern[i:i+k]
        l.append(s)
        
    return l
```

```python
k = int(input())
pattern = input()

l = get_all_k_mers(pattern, k)
l.sort()

print(('\n').join(l))
```

# Problem BA3B

```python
def reconstruct(dna):
    ans = ""
    ans += dna[0]
    for i in range(1,len(dna)):
        x = dna[i]
        ans += x[-1]
    return ans
```

```python
dna = input()
dna = dna.split()

ans = reconstruct(dna)
print(ans)
```

# Problem BA3C

```python
def overlap_graph(dna):
    adj = [[0 for x in range(len(dna))] for y in range(len(dna))]
    for i in range(len(dna)):
        for j in range(len(dna)):
            if i!=j:
                if dna[i][1:] == dna[j][:-1]:
                    adj[i][j] = 1
    return adj
```

```python
dna = input()
dna = dna.split()

adj = overlap_graph(dna)

for i in range(len(adj)):
    for j in range(len(adj[i])):
        if(adj[i][j]==1):
            print(dna[i] + " -> " + dna[j])
```

# Problem BA3D

```python
def DeBruijn(dna, k):
    adj = {}
    for i in range(0, len(dna)-k+1):
        string = dna[i: i+k]
        if(string[0:-1] in adj):
            adj[string[0:-1]].append(string[1:])
        else:
            adj[string[0:-1]] = []
            adj[string[0:-1]].append(string[1:])
    return adj
```

```python
k = int(input())
dna = input()

adj = DeBruijn(dna, k)

for keys, values in adj.items():
    print(keys, end=' -> ')
    print((','.join(values)))
```

# Problem BA3E

```python
def DeBruijn_again(dna):
    adj = {}
    for string in dna:
        x = string[0: -1]
        y = string[1: ]
        if(x in adj):
            adj[x].append(y)
        else:
            adj[x] = []
            adj[x].append(y)
    return adj
```

```python
dna = input()
dna = dna.split()

adj = DeBruijn_again(dna)

for keys, values in adj.items():
    print(keys, end=' -> ')
    print((','.join(values)))
```
